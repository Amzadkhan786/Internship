```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
#Fine The Best Random State For Our Model
from sklearn.linear_model import LinearRegression
max_accuracy = 0
best_rm = 0
for i in range(0,2000):
    x_train,x_test,y_train,y_test = train_test_split(X,y,test_size =20,random_state = i)
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    lr_pred_train = lr.predict(x_train)
    lr_pred = lr.predict(x_test)
    accuracy = r2_score(y_test,lr_pred)
    if accuracy> max_accuracy:
        max_accuracy =accuracy
        best_rm = i
print("Maximum R2 Score:-",max_accuracy,"In This Best Random State:-",best_rm)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size =20,random_state = best_rm)
```

# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data

## This task will be done in following steps:
#### 1. First get the webpage https://www.shine.com/
#### 2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
#### 3. Then click the searchbutton.
##### 4. Then scrape the data for the first 10 jobs results you get.
#### 5. Finally create a dataframe of the scraped data.

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.shine.com/")
```

```python
#Click On Job Button Then Click On It
on_search_bar = driver.find_element(By.XPATH,"/html/body/div/header/div[3]/div/div/div[1]/div/span")
on_search_bar.click()
```

```python
# Fill Job Title 
job_title = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input")
job_title.send_keys('Data Analyst')
```

```python
#Fill Location 
location = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')
```

```python
# Then Search 
search = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
search.click()
```

```python
#Make Empty List [] For Storing Data
job_title = []
job_location = []
company_name = []
job_experience = []


#Title Tag find
t_tag = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')  #Title Tag Xpath
for i in t_tag[:10]:   #  USing For Loop    
    title = i.text     #Initiate in Title
    job_title.append(title)  # Stor In Job_ Title List

#Location Find 
loc_tag = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for loc in loc_tag[:10]: #Using For Loop  [:10] Bcoz For 10 Jobs Data
    location = loc.text   # Initiate In Location
    job_location.append(location)  # Store In Job Location List  variable

#Find Company
company = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company[:10]: #Using For Loop
    name = i.text     #Initiate in Name
    company_name.append(name.title()) # Store In Company NAme

#Find Experience Data
experience = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for ex in experience[:10]: # Using Loop
    exp = ex.text      #Initiate In Exp
    job_experience.append(exp)  #Store in Job Experience List 
```

# DataFrame

```python
data = pd.DataFrame({'Job Title': job_title,'Location':job_location,'Company':company_name,'Experience':job_experience})
data
```

```python
#Scrap Product Ratings
product_rating = []  #Create Empty List 
for page in range(0,10):#Set Pages Range  For 10 pages
    print('Scraping Page',page+1)
    rating_tag = driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]') # Rating Xpath Initiate In Rating Tag
    for i in rating_tag:   #Using For Loop
        product_rating.append(i.text) # Import Data In Empty List
    next_button = driver.find_element(By.XPATH,'//nav[@class="yFHi8N"]/a') # Next Button XPATH intiate In 'next_button' Variable
    next_button.click()  #Click On it
    time.sleep(3)   # Wait For 3 Second When Going Next Page

```

```python
print(product_rating)
```

```python
# Scrap Review
review_summary = []    #Create Empty List
for page in range(0,10):   #Using Loop For Scrap 10 Pages
    print('Scraping Page',page+1)     #Showing pages
    review_tag = driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')   #XPATH
    for i in review_tag:
        review_summary.append(i.text)   #Import Data In Empty List
    next_button = driver.find_element(By.XPATH,'//nav[@class="yFHi8N"]/a')  # Next Button
    next_button.click()         #Click
    time.sleep(3)   # Wait For 2 Sec.
    
```

```python
#Scraping Full Review
full_review = []       
for page in range(0,10):   #Set range For 10 Pages
    print('Scraping Page',page+1)  #Showing Scraping Pages
    f_review_tag = driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]') #XPATH
    for i in f_review_tag:
        full_review.append(i.text)   #Import Data In Empty List
    next_button = driver.find_element(By.XPATH,'//nav[@class="yFHi8N"]/a')  #XPATH for Next Button
    next_button.click()  #Click On It
    time.sleep(2)
```

```python
#Scrap Aurthors Name 
aurthors_name = []  #Create  Empty List

for page in range(0,10): # Using For Loop  And Set Range For 10 Pages
    print("Scraping_page",page+1)   # Showing Scraping Pages
    aurthors_tag = driver.find_elements(By.XPATH,'//div[@class="author"]')  # XPATH
    for i in aurthors_tag:  # Using Loop For Data
        aurthors_name.append(i.text)   # Import Data In Empty List
    next_page = driver.find_element(By.XPATH,'//div[@class="pager"]')  #Next Page XPATH
    next_page.click()  #Click On Next Page
    time.sleep(2)   # 2 Second Wait For Page Loading
```
